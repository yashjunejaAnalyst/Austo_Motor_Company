**QUES 1 ) Do men tend to prefer SUVs more compared to women ?**

```python
sns.countplot(data=AustoMotor,x='Make', hue='Gender')
plt.xticks(rotation=90)
```

**QUES 2.) What is the Likelihood of a salaried person buying a Sedan ?**

```python
sns.countplot(data=AustoMotor,x='Make', hue='Profession')
plt.xticks(rotation=90)
```

**QUES 3.) What Evidence or data supports Sheldon Cooper’s Claim that a salaried male is an easier target for a SUV sale over a Sedan sale ?**

```python
sns.countplot(data=AustoMotor,x='Make', hue='Gender')
plt.xticks(rotation=90)
```

```python
a=sns.FacetGrid(data=AustoMotor, col='Profession', hue='Gender')
a.map(plt.scatter,'Make','Age')
a.add_legend()
plt.show()
```
